## **Implementación del código**

```python
# Datos proporcionados
X = np.array([50, 52, 54, 56, 58, 60, 62])  # Valores de X
Y = np.array([31, 50, 77, 96, 112, 130, 162])  # Valores de Y

# Calcular las sumas necesarias para el sistema de ecuaciones
n = len(X)  # Número de puntos de datos
sum_x = np.sum(X)  # Suma de todos los valores de X
sum_y = np.sum(Y)  # Suma de todos los valores de Y
sum_x2 = np.sum(X**2)  # Suma de X al cuadrado
sum_x3 = np.sum(X**3)  # Suma de X al cubo
sum_x4 = np.sum(X**4)  # Suma de X a la cuarta potencia
sum_xy = np.sum(X * Y)  # Suma del producto de X y Y
sum_x2y = np.sum(X**2 * Y)  # Suma del producto de X al cuadrado y Y

# Construir el sistema de ecuaciones para resolver los coeficientes a, b y c
A = np.array([[sum_x4, sum_x3, sum_x2],  # Matriz del sistema de ecuaciones
              [sum_x3, sum_x2, sum_x],
              [sum_x2, sum_x, n]])

B = np.array([sum_x2y, sum_xy, sum_y])  # Vector de términos independientes

# Resolver el sistema de ecuaciones lineales para encontrar los coeficientes a, b, c
coefficients = np.linalg.solve(A, B)  # Resolver el sistema usando álgebra lineal de numpy
a, b, c = coefficients  # Asignar los valores de los coeficientes

# Mostrar los coeficientes encontrados
print(f"Coeficiente a: {a}")  # Imprimir el coeficiente 'a'
print(f"Coeficiente b: {b}")  # Imprimir el coeficiente 'b'
print(f"Coeficiente c: {c}")  # Imprimir el coeficiente 'c'

# Calcular el valor de Y para X = 64 usando los coeficientes obtenidos
X_new = 64  # Nuevo valor de X para el cual queremos predecir Y
Y_pred = a * X_new**2 + b * X_new + c  # Usar la ecuación cuadrática para predecir Y
print(f"El valor predicho de Y para X=64 es: {Y_pred}")  # Imprimir el valor predicho de Y

```

```python
# Datos proporcionados
X = np.array([50, 52, 54, 56, 58, 60, 62])  # Valores de X
Y = np.array([31, 50, 77, 96, 112, 130, 162])  # Valores de Y correspondientes

# Resolver el sistema de ecuaciones para los coeficientes a, b, c usando regresión cuadrática
A = np.vstack([X**2, X, np.ones(len(X))]).T  # Construir la matriz del sistema de ecuaciones
coefficients = np.linalg.lstsq(A, Y, rcond=None)[0]  # Resolver el sistema para encontrar los coeficientes
a, b, c = coefficients  # Asignar los coeficientes a, b y c

# Generar valores de X para la gráfica, entre el mínimo y máximo de los datos originales
X_plot = np.linspace(X.min(), X.max(), 100)  # Crear 100 puntos equidistantes entre el valor mínimo y máximo de X
Y_plot = a * X_plot**2 + b * X_plot + c  # Calcular los valores correspondientes de Y usando el modelo cuadrático

# Calcular y mostrar el valor predicho de Y para X = 64
X_new = 64  # Nuevo valor de X para el cual queremos predecir Y
Y_pred = a * X_new**2 + b * X_new + c  # Usar la ecuación cuadrática para calcular Y

# Graficar los puntos de datos originales, la curva de ajuste cuadrático y la predicción
plt.scatter(X, Y, color='blue', label='Datos originales')  # Grafica los datos originales como puntos azules.
plt.plot(X_plot, Y_plot, color='red', label=f'Ajuste cuadrático: $Y = {a:.2e}X^2 + {b:.2e}X + {c:.2e}$')  # Grafica la curva de ajuste en rojo.
plt.scatter(X_new, Y_pred, color='green', label=f'Predicción para $X=64$: {Y_pred:.2f}')  # Grafica la predicción para X=64 como un punto verde.

# Etiquetas de los ejes
plt.xlabel('X')  # Etiqueta del eje X.
plt.ylabel('Y')  # Etiqueta del eje Y.

# Añadir leyenda y título
plt.legend()  # Muestra la leyenda para diferenciar entre datos originales, ajuste y predicción.
plt.title('Ajuste Cuadrático')  # Título del gráfico.
plt.title('GRUPO 6', loc="left")
plt.title('Fabricio Baquero', loc="right")

# Mostrar grillas
plt.grid(True, which='both', linestyle='--', linewidth=0.7)  # Activa grillas mayores y menores con líneas punteadas.

# Mostrar gráfico
plt.show()  # Muestra el gráfico final.
```
